```python
def PlotBarCharts(inpData, colsToPlot, title = "Grafico de barras de", colors = ['red']):
    %matplotlib inline
    
    # Gerando múltiplos subplots
    fig, subPlot = plt.subplots(nrows = 1, ncols = len(colsToPlot), figsize = (20, 5))
    fig.suptitle(title, fontsize=16)  # Removido o str(colsToPlot)

    if len(colsToPlot) == 1:  # Se houver apenas um gráfico de barras
        colName = colsToPlot[0]
        # Plotando gráfico de barras
        counts = inpData[colName].value_counts()
        counts.plot(kind ='bar', ax = subPlot, color = colors[0])  # Aplique a primeira cor da lista
        # Definindo os rótulos personalizados para utilizar no eixo x
        subPlot.set_xticklabels(counts.index, rotation = 45, ha = 'right')
        subPlot.legend([colName])  # Adicionando as legendas
    else:
        for colName, ax, color in zip(colsToPlot, subPlot, colors):  # Intera sobre as cores
            # Plotando gráfico de barras
            counts = inpData[colName].value_counts()
            counts.plot(kind = 'bar', ax = ax, color = color)
            # Definindo os rótulos personalizados para no eixo x
            ax.set_xticklabels(counts.index, rotation = 45, ha = 'right')
            ax.legend([colName])  # Adicionando as legendas
```

```python
def identify_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

def plot_boxplot(data, column):
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=data[column])
    plt.title(f'Boxplot de {column}')
    plt.show()

# Variáveis a serem analisadas
numeric_columns = ['credit_duration', 'credit_amount', 'age', 'entry_payment']

for column in numeric_columns:
    outliers = identify_outliers(data, column)
    plot_boxplot(data, column)
    print(f"Outliers para {column}:")
    print(outliers)
```

<font size= 6 fonte color='navy'> <b> Teste Anova
 </font>

Teste ANOVA é como um canivete suíço para comparar médias entre três ou mais grupos diferentes. Basicamente, ele nos ajuda a descobrir se há uma diferença real entre os grupos ou se é apenas coisa da nossa cabeça.

Quando você tem um monte de grupos diferentes e quer saber se alguma diferença entre eles é significativa, é aí que o ANOVA entra em cena. É como dizer: "Ei, pessoal, esses grupos são realmente diferentes ou é só impressão minha?"

Então, se você está fazendo um experimento com várias condições ou quer entender como diferentes fatores afetam uma variável, o ANOVA é a ferramenta que você precisa. Ele ajuda a responder perguntas como: "O efeito do tipo de tratamento na eficácia do medicamento é realmente diferente ou é só coincidência?"

Resumindo, o ANOVA é como o árbitro de uma luta entre grupos diferentes de dados. Ele decide se há um verdadeiro vencedor ou se é apenas um empate.


```python
# Definição de uma função para encontrar a relação estatística com todas as variáveis categóricas
def funcao_anova(inpData, variavel_target, lista_preditores_continuos):
    
# Criação de uma lista vazia de preditores finais selecionados
    preditores_selecionados = []
    
    print('##### Resultado da ANOVA ##### \n')
    for predictor in lista_preditores_continuos :
        listas_groupo_categoria = inpData.groupby(variavel_target)[predictor].apply(list)
        anova_resultado = f_oneway(*listas_groupo_categoria)
        
# Condição o P-valor da ANOVA for <0,05, rejeita-se H0
        if (anova_resultado[1] < 0.05):
            print(predictor, 'está correlacionado com', variavel_target, '| P-Valor:', anova_resultado[1])
            preditores_selecionados.append(predictor)
        else:
            print(predictor, 'não está correlacionado com', variavel_target, '| P-Valor:', anova_resultado[1])
    
    return preditores_selecionados
```

```python
# Chamada da função para verificar quais variáveis categóricas estão correlacionadas com a Target
ContinuousVariables = ['credit_duration', 'credit_amount', 'age', 'entry_payment']
funcao_anova(inpData = data, variavel_target = 'credit_risk', lista_preditores_continuos = ContinuousVariables)
```

Você já se perguntou como os cientistas de dados descobrem as conexões entre diferentes categorias de dados? Bem, uma ferramenta importante que usamos para isso é o **teste do qui-quadrado**, também conhecido como **Chi-Square Test**.

O que torna esse teste tão útil é que ele nos ajuda a ver se existe uma relação significativa entre duas categorias de dados. Tal como no presente _case_, que estamos analisando o credit_risk, podemos usar o teste do qui-quadrado para ver se há uma associação entre a idade dos clientes e oo risco de crédito e assim por diante.

Ao fazer isso, podemos descobrir padrões interessantes, como se os clientes mais jovens têm um risco maior ou não, enquanto os clientes mais velhos seriam mais robustos quanto ao crédito? Essas informações são super úteis para empresas, pois podem ajudá-las a personalizar suas estratégias de marketing e comercial, delineando cenarios de riscos calculados. 
Portanto se trata de uma ferramenta poderosa que nos ajuda a entender melhor o mundo ao nosso redor, descobrindo padrões e relações nos dados que podem passar despercebidos de outra forma.


```python
# Esta função calcula a correlação entre todas as variáveis categóricas e a variável alvo
def funcao_chisq(inpData, variavel_target, lista_variaveis_categoricas):
    from scipy.stats import chi2_contingency
    
    
    # Criação de uma lista vazia de preditores finais selecionados
    preditores_selecionados = []

    for predictor in lista_variaveis_categoricas:
        resultado_cross_tab = pd.crosstab(index = inpData[variavel_target], columns=inpData[predictor])
        resultado_chisq = chi2_contingency(resultado_cross_tab)
        
        # Se ChiSq P-Valor is <0.05, rejeita-se H0
        if (resultado_chisq[1] < 0.05):
            print(predictor, 'está correlacionado com', variavel_target, '| P-Valor:', resultado_chisq[1])
            preditores_selecionados.append(predictor)
        else:
            print(predictor, 'não está correlacionado com', variavel_target, '| P-Valor:', resultado_chisq[1])        
            
    return(preditores_selecionados)
```

```python
# Lista de variáveis categóricas
variaveis_categoricas = ['account_status', 'history_of_compliance', 'credit_purpose', 'savings', 'employment_duration', 'other_installment_plans', 
                    'installment_rate', 'personal_status_sex', 'other_debtors', 'present_residence', 'property', 'type_of_housing', 
                    'number_credits', 'job', 'level_of_education', 'foreign_worker', 'people_liable', 'telephone']

# Calling the function
funcao_chisq(inpData = data, 
              variavel_target = 'credit_risk',
              lista_variaveis_categoricas = variaveis_categoricas)
```

# Utilização da função `get_dummies()` no German Credit Risk

No contexto do German Credit Risk, a utilização da função `get_dummies()` pode oferecer várias vantagens ao lidar com variáveis categóricas:

1. **Melhora o desempenho de modelos de aprendizado de máquina:**
   Muitos modelos de aprendizado de máquina, como Árvores de Decisão, Regressão Linear e Gradient Boosting, funcionam melhor com dados numéricos. Ao converter variáveis categóricas em vetores de dummies, cria-se representações numéricas que esses modelos podem facilmente processar e utilizar para aprender relacionamentos complexos entre as variáveis e a classe a ser predita.

2. **Captura mais informações:**
   Ao utilizar `get_dummies()`, cria-se uma coluna separada para cada possível categoria dentro da variável categórica original. Isso permite ao modelo capturar as nuances e diferenças entre as categorias de forma mais granular, em vez de tratá-las como um único valor.

3. **Evita codificação ordinal errônea:**
   Codificar categorias como números ordinais (ex: Solteiro=1, Casado=2, Divorciado=3) pode ser um erro, pois isso implica uma ordem que pode não existir na realidade (divórcio não é necessariamente "mais importante" que ser solteiro). `get_dummies()` evita esse problema criando representações binárias independentes para cada categoria.

# ML: Dividindo os Dados em Amostras de Treinamento e Teste

No processo de construção de modelos de aprendizado de máquina, não utilizamos todos os dados disponíveis. Em vez disso, parte dos dados é reservada para avaliar a qualidade do modelo. Essa divisão dos dados é fundamental para garantir que o modelo seja capaz de generalizar bem para novos dados.

## Dados de Treinamento e Teste

Os dados são divididos em duas partes principais:

- **Dados de Treinamento:** Essa parte dos dados, geralmente representando cerca de 70% do conjunto total, é usada para treinar o modelo. Durante o treinamento, o modelo aprende padrões e relações nos dados que serão usados para fazer previsões futuras.
  
- **Dados de Teste:** Os dados de teste consistem nos 30% restantes do conjunto total. Esses dados são mantidos separados durante o treinamento e são usados para avaliar o desempenho do modelo após o treinamento. Essa avaliação nos dá uma medida de quão bem o modelo é capaz de generalizar para dados não vistos anteriormente.

Essa prática ajuda a evitar o sobreajuste (overfitting), onde o modelo se ajusta muito bem aos dados de treinamento, mas não é capaz de generalizar para novos dados. Ao avaliar o desempenho do modelo em dados de teste independentes, podemos obter uma estimativa mais realista de sua capacidade de fazer previsões precisas em situações do mundo real.


```python
dados_num.columns
```

```python
# Definição das variáveis independentes e dependentes
variavel_target = 'credit_risk'
outras_variaveis = [col for col in dados_num.columns if col != variavel_target]

# Extraindo os valores das variáveis independentes e dependentes
X = dados_num[outras_variaveis].values
y = dados_num[variavel_target].values
```

# Normalização de dados

```python
# Dividir os dados em conjuntos de treinamento e teste
# X_train: conjunto de treinamento das variáveis independentes
# X_test: conjunto de teste das variáveis independentes
# y_train: conjunto de treinamento da variável dependente
# y_test: conjunto de teste da variável dependente
# test_size: proporção do conjunto de teste (30% neste caso)
# random_state: semente para garantir reprodutibilidade dos resultados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=428)
```

 <font size= 5 fonte color='navy'> <b> Uso da Regressão Logistica
 </font>

A **regressão logística** emerge como uma opção sólida para lidar com o German Credit Risk, e há várias razões para isso:

1. **Simplicidade:**
   A regressão logística é um modelo relativamente simples de entender e implementar, tornando-a ideal para iniciantes ou para aqueles que buscam uma solução rápida e de fácil interpretação.
   Sua simplicidade facilita a compreensão do relacionamento entre as variáveis e a probabilidade de inadimplência, tornando-a útil para comunicação e análise.

2. **Interpretabilidade:**
   A regressão logística fornece coeficientes para cada variável, os quais indicam a força e a direção de seu impacto na probabilidade de inadimplência.
   Essa interpretabilidade é crucial no contexto do German Credit Risk, pois permite identificar os fatores que mais influenciam a inadimplência e tomar decisões mais informadas sobre a concessão de crédito.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, f1_score

# Inicializando o classificador Logistic Regression
clf = LogisticRegression(C=1, penalty='l2', solver='newton-cg')

# Treinando o modelo
LOG = clf.fit(X_train, y_train)

# Fazendo previsões no conjunto de teste
predictions = LOG.predict(X_test)

# Avaliando o desempenho do modelo
print("Relatório de Classificação:")
print(classification_report(y_test, predictions))
print("Matriz de Confusão:")
print(confusion_matrix(y_test, predictions))

# Calculando a acurácia do modelo
accuracy = clf.score(X_test, y_test)
print('Acurácia do modelo no conjunto de teste:', round(accuracy, 2))

# Calculando o F1-Score
f1 = f1_score(y_test, predictions, average='weighted')
print('F1-Score do modelo no conjunto de teste:', round(f1, 2))

# Executando a validação cruzada de 10-Fold para calcular a média da acurácia
accuracy_values = cross_val_score(clf, X, y, cv=10, scoring='f1_weighted')
print('\nValores de acurácia para validação cruzada de 10-Fold:\n', accuracy_values)
print('\nAcurácia média final do modelo:', round(accuracy_values.mean(), 2))
```
